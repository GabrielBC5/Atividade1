```python
# 1. Peça 5 números com for e diga quantos são pares.

quantidade_pares = 0

for i in range(5):
    numero = int(input(f"Digite o {i+1}º número: "))
    
    if numero % 2 == 0:
        quantidade_pares += 1

print("Quantidade de números pares:", quantidade_pares)
```

```python
# 2. Use while para pedir uma nota entre 0 e 10.

nota = -1

while nota < 0 or nota > 10:
    nota = float(input("Digite uma nota entre 0 e 10: "))

print("Nota válida:", nota)
```

```python
# 3. Peça um número e mostre a tabuada com for.

numero = int(input("Digite um número para ver a tabuada: "))

for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

```python
# 4. Peça idade e documento. Use if com and para verificar acesso.

idade = int(input("Digite sua idade: "))
documento = input("Você tem documento? (sim/não): ").strip().lower()

if idade >= 18 and documento == "sim":
    print("Acesso permitido.")
else:
    print("Acesso negado.")
```

```python
# 5. Peça uma palavra e mostre letra por letra com for.

palavra = input("Digite uma palavra: ")

for letra in palavra:
    print(letra)
```

```python
# 6. Crie um sistema de login com 3 tentativas usando while.

usuario_correto = "admin"
senha_correta = "1234"

tentativas = 0
acesso = False

while tentativas < 3 and not acesso:
    usuario = input("Digite o usuário: ")
    senha = input("Digite a senha: ")
    
    if usuario == usuario_correto and senha == senha_correta:
        print("Login realizado com sucesso!")
        acesso = True
    else:
        tentativas += 1
        print("Usuário ou senha incorretos.")
        print("Tentativas restantes:", 3 - tentativas)

if not acesso:
    print("Número máximo de tentativas atingido.")
```
