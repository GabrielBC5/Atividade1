```python
# 1. Peça a idade e se a pessoa tem carteira de motorista. Mostre se ela pode dirigir.

idade = int(input("Digite sua idade: "))
carteira = input("Você tem carteira de motorista? (sim/não): ").strip().lower()

if idade >= 18 and carteira == "sim":
    print("Você pode dirigir.")
else:
    print("Você não pode dirigir.")
```

```python
# 2. Peça um número. Verifique se ele está entre 50 e 100.

numero = float(input("Digite um número: "))

if 50 <= numero <= 100:
    print("O número está entre 50 e 100.")
else:
    print("O número não está entre 50 e 100.")
```

```python
# 3. Peça o dia da semana. Mostre se é final de semana.

dia = input("Digite o dia da semana: ").strip().lower()

if dia == "sábado" or dia == "sabado" or dia == "domingo":
    print("É final de semana.")
else:
    print("Não é final de semana.")
```

```python
# 4. Peça usuário e senha. Verifique se ambos estão corretos.

usuario_correto = "admin"
senha_correta = "1234"

usuario = input("Digite o usuário: ")
senha = input("Digite a senha: ")

if usuario == usuario_correto and senha == senha_correta:
    print("Login correto.")
else:
    print("Usuário ou senha incorretos.")
```

```python
# 5. Peça uma nota e a frequência do aluno. Mostre se ele foi aprovado.

nota = float(input("Digite a nota do aluno: "))
frequencia = float(input("Digite a frequência do aluno em %: "))

if nota >= 7 and frequencia >= 75:
    print("Aluno aprovado.")
else:
    print("Aluno reprovado.")
```

```python
# 6. Crie um menu que só encerra quando o usuário digitar 0

opcao = -1

while opcao != 0:
    print("\nMenu:")
    print("1 - Opção 1")
    print("2 - Opção 2")
    print("3 - Opção 3")
    print("0 - Sair")
    
    opcao = int(input("Escolha uma opção: "))
    
    if opcao == 1:
        print("Você escolheu a Opção 1.")
    elif opcao == 2:
        print("Você escolheu a Opção 2.")
    elif opcao == 3:
        print("Você escolheu a Opção 3.")
    elif opcao == 0:
        print("Encerrando o programa...")
    else:
        print("Opção inválida.")
```

```python
# 1. Peça 5 números com for e diga quantos são pares.

quantidade_pares = 0

for i in range(5):
    numero = int(input(f"Digite o {i+1}º número: "))
    
    if numero % 2 == 0:
        quantidade_pares += 1

print("Quantidade de números pares:", quantidade_pares)
```

```python
# 6. Crie um sistema de login com 3 tentativas usando while.

usuario_correto = "admin"
senha_correta = "1234"

tentativas = 0
acesso = False

while tentativas < 3 and not acesso:
    usuario = input("Digite o usuário: ")
    senha = input("Digite a senha: ")
    
    if usuario == usuario_correto and senha == senha_correta:
        print("Login realizado com sucesso!")
        acesso = True
    else:
        tentativas += 1
        print("Usuário ou senha incorretos.")
        print("Tentativas restantes:", 3 - tentativas)

if not acesso:
    print("Número máximo de tentativas atingido.")
```
